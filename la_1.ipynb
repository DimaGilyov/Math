__1.__ Исследовать на линейную зависимость:

$$f_{1}(x) = e^{x}, f_{2}(x) = 1, f_{3}(x) = x + 1, f_{4}(x) = x - e^{x}.$$


Рассмотрим линейную комбинацию этих функций: $\alpha_{1} \cdot f_{1}(x) + \alpha_{2} \cdot f_{2}(x) + \alpha_{3} \cdot f_{3}(x) + \alpha_{4} \cdot f_{4}(x)$ 

Если ∀x ∈ R равенство $\alpha_{1} \cdot f_{1}(x) + \alpha_{2} \cdot f_{2}(x) + \alpha_{3} \cdot f_{3}(x) + \alpha_{4} \cdot f_{4}(x) = 0$ выполняется только при $\alpha_{1} = \alpha_{2} = \alpha_{3} = \alpha_{4} = 0$, то рассматриваемые функции линейно независимы.

Если ∀x ∈ R равенство $\alpha_{1} \cdot f_{1}(x) + \alpha_{2} \cdot f_{2}(x) + \alpha_{3} \cdot f_{3}(x) + \alpha_{4} \cdot f_{4}(x) = 0$ возможно при условии, что хотя бы один из коэффициентов $\alpha_{i}$ не равен нулю, то функции линейно зависимы.

Подставим в выражение заданные функции:

$\alpha_{1} \cdot e^{x} + \alpha_{2} \cdot 1 + \alpha_{3} \cdot (x + 1) + \alpha_{4} \cdot (x-e^{x}) = 0$

Раскрываем скобки и перегруппируем слагаемые:

$(\alpha_{1} - \alpha_{4}) \cdot e^{x} + (\alpha_{3} + \alpha_{4}) \cdot х + (\alpha_{2} + \alpha_{3})) = 0$

Последнее равенство возможно лишь в том случае, когда коэффициенты одновременно равны нулю, т.е.:

$\begin{cases} \alpha_{1} - \alpha_{4} = 0 \\\ \alpha_{3} + \alpha_{4} = 0\ \\\ \alpha_{2} + \alpha_{3} = 0\end{cases}$

Мы получили систему линейных уравнений. Если решение лишь одно – нулевое, $\alpha_{1} = \alpha_{2} = \alpha_{3} = \alpha_{4} = 0$, то функции линейно независимы.

Если же есть иные решения, кроме нулевого, то функции линейно зависимы.

Можно рассуждать коротко, если один из векторов $x_{1}, x_{2}, x_{3}, ...$ линейно выражается через остальные. Например,

$x_{1} = \alpha_{2} \cdot x_{2} + \alpha_{3} \cdot x_{3} + ... +\alpha_{n} \cdot x_{n}$

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

Cпособы определения нормы.

- $p$-норма (норма Гёльдера) для вектора $x = (x_{1}, \dots, x_{n}) \in \mathbb{R}^{n}$ вычисляется по формуле:

$$\left\| x \right\|_{p} = \left( \sum_{i} \left| x_{i} \right|^{p} \right)^{1 / p},~p \geq 1.$$

В частных случаях при:

- $p = 1$ получаем $\ell_{1}$ норму (_манхэттенская норма_, или _манхэттенское расстояние_)

$$\left\|x\right\|_{1} = \sum_{i}|x_{i}|;$$

- $p = 2$ получаем $\ell_{2}$ норму (_евклидова норма_, наиболее часто используемая)

$$\left\|x\right\|_{2} = \sqrt{\sum_{i}|x_{i}|^{2}}.$$

В любом вещественном евклидовом пространстве можно ввести понятие _угла_ между двумя произвольными элементами $x$ и $y$. Углом $\varphi \in [0, \pi]$ между этими элементами назовем угол, косинус которого определяется соотношением

$$cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$$

Найдем нормы векторов $x = (4, 2, 4)$,  $y = (12, 3, 4)$

Манхэттенская норма вектора $x$:

$$\left\|x\right\|_{1} = |4|+|2|+|4| = 10 $$

Манхэттенская норма вектора $y$:

$$\left\|x\right\|_{1} = |12|+|3|+|4| = 19 $$


Евклидова норма вектора $x$:

$$\left\|x\right\|_{2} = \sqrt{4^{2} + 2^{2} + 4^{2}} = \sqrt{36} = 6$$


Евклидова норма вектора $y$:

$$\left\|x\right\|_{2} = \sqrt{12^{2} + 3^{2} + 4^{2}} = \sqrt{169} = 13$$

Угол между векторами $x = (4, 2, 4)$,  $y = (12, 3, 4)$:

$$cos\varphi = \frac{4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4}{\sqrt{4^{2} + 2^{2} + 4^{2}}\sqrt{12^{2} + 3^{2} + 4^{2}}} = \frac{70}{\sqrt{36} \cdot \sqrt{169}} = \frac{70}{6 \cdot 13}\approx 0.897$$

```python
from numpy.linalg import norm
import math
```

```python
x = np.array([4, 2, 4])

y = np.array([12, 3, 4])

print(f'l1 Манхетовская норма вектора x: {norm(x, ord=1)}\n')

print(f'l1 Манхетовская норма вектора y: {norm(y, ord=1)}\n')

print(f'l2 Евклидова норма вектора x: {norm(x, ord=2)}\n')

print(f'l2 Евклидова норма вектора y: {norm(y, ord=2)}\n')

cos_phi = np.dot(x, y) / norm(x, ord=2) / norm(y, ord=2)

print(f'cosφ ≈ {round(cos_phi,3)} => φ ≈ 26° \n')


phi = math.acos(cos_phi)

print(f'Угол между векторами x и y в радианах: {round(phi,2)}\n')

print(f'Угол между векторами x и y в градусах: {round(math.degrees(phi))}°\n')

#degrees(rad)
#arccos(0,897) в градусах = 26.2335
#arccos(0,897) в радианах = 0.4579
```
